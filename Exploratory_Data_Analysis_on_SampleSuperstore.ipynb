```
# First we will import important libraries required in this notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# we will import our dataset
data = pd.read_csv("SampleSuperstore.csv")
# we will print first five line of our dataset
data.head()
```

**Performing Exploratory Data Analysis**

```
# we will check if there is any NAN value in our dataset or not
data.isnull().sum()
```

```
# we will check shape of our dataset
data.shape
```

```
# we will print the name of columns of ours dataset
data.columns.values
```

What are total sales and profit of US company.

```
# We will calculate sales of company
sale = data["Sales"].sum()
sale
```

```
# we will calculate profit of company
profit = data["Profit"].sum()
profit
```

Top 10 sales and profit of US company

```
# we will print top 10 sales of company
Top_10_sales = data.groupby("State").Sales.sum().nlargest(n=10)
Top_10_sales
```

```
# we will print top 10 profir of company
Top_10_profit = data.groupby("State").Profit.sum().nlargest(n=10)
Top_10_profit
```

```
# we will plot bar graph of top 10 sales of company

plt.style.use('dark_background')
Top_10_sales.plot(kind ='bar',color="blue",figsize =(10,6), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Top 10 States by Sales",fontsize =16)
plt.show()
```

```
# we will plot bar graph of top 10 profit of company

plt.style.use('dark_background')
Top_10_profit.plot(kind ='bar',color="blue", figsize =(10,6), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Profit",fontsize =13)
plt.title("Top 10 States by Profit",fontsize =16)
plt.show()
```

```
# we will plot correlation graph
sns.heatmap(data.corr(),annot=True)
```

from above we can see that sales and profit are positively correlated and discount and profit are negatively correlated.
